```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create symulation function

```python
def f(x_cent, y_cent, d, angle, sigma, N_samples=100):
    samples = np.random.multivariate_normal([x_cent,y_cent],sigma*np.eye(2),size=N_samples)
    shifts = np.random.uniform(low=-d/2., high=d/2. ,size=N_samples)
    shifts = np.outer(shifts,np.array([np.cos(angle), np.sin(angle)]))
    return shifts + samples
```

```python
def ff(x_cent, y_cent, d, angle, sigma, N_samples=3):
    samples = int(10**N_samples)
#     print(N_samples)
    p=f(x_cent, y_cent, d, angle, sigma, N_samples=samples)
    image,_,_=np.histogram2d(p[:,0],p[:,1],bins = [np.arange(-10,11), np.arange(-10,11)])
    return image
```

```python
%time p = f(0, 0, 4, angle=0, sigma=1, N_samples=10000)
image,_,_=np.histogram2d(p[:,0],p[:,1],bins = [np.arange(-10,11), np.arange(-10,11)])
plt.imshow(image)
```

## Read picture

```python
from PIL import Image
def read_image(file_name):
    image = Image.open(file_name)
    image = np.sum(image, axis=2)
    return image.astype(np.float64)
```

```python
image = read_image('example_data/longer_exposure.JPG')
image -= np.median(image)
```

```python
plt.imshow(image)
```

```python
x = 1000
y = 988
d = 10
p = ff(0, 0, 4, angle=0, sigma=1, N_samples=2)
print(image[x-d:x+d,y-d:y+d].shape)


from scipy.optimize import minimize

def fff(x):
    return ff(x[0], x[1], x[2], angle=x[3], sigma=x[4], N_samples=x[5])

def chi2(x,data):
    
    p = fff(x)
    dd = (data - p).flatten()
    return dd.dot(dd.T)

x0=(1,2,4,0,3,4)
# chi2(x0,image[x-d:x+d,y-d:y+d])
res = minimize(chi2, x0, args=(image[x-d:x+d,y-d:y+d],),
              bounds = [(-5,5),(-5,5),(0,10),(0,np.pi),(0,5),(0,6)],
              method = 'Nelder-Mead',
              options={'maxiter':5000})
print(res)
plt.imshow(image[x-d:x+d,y-d:y+d] - fff(res.x))
plt.figure()
plt.imshow(fff(res.x))
```
